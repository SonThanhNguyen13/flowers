```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
while True:
    if len(X) >= 20:
        X.sort()
        break
    else:
        x = random.randint(30,60)
        if x not in X:
            X.append(x)
```

```python
while True:
    if len(Y) >= 20:
        Y.sort()
        break
    else:
        y = random.randint(50,100)
        if y not in Y:
            Y.append(y)
```

```python
len(Y)
```

```python
plt.scatter(X,Y)
```

```python
w = np.array((3,4))
```

```python
X1 = [[i, 1] for i in X]
```

```python
X1 = np.asarray(X1)
```

```python
#plt.figure(figsize=(20,5))
plt.scatter(X, Y)
loss_prev = 0
while True:
    pre = np.dot(X1, w)
    plt.plot(X, pre)
    loss = np.sum((1 / 2) * np.square(pre - Y))
    print(loss)
    if (abs(loss - loss_prev)) < 1:
        break
    loss_prev = loss
    da = np.sum(np.multiply((pre - Y),X))
    db = np.sum(pre - Y)
    d = np.array((da,db))
    w = w - 0.00001 * d
```
